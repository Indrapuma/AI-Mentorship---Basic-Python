```python
import datetime

#pengantar
print("Hai aku AIs, asisten virtual pertamamu\n")
print("siapa namamu ?")

nama = str(input("Masukkan Nama : "))

print("\nSalam kenal{}".format(nama))
print("Apakah ada yang bisa saya bantu?")

#Menu utama
def show_menu():
    print("\nSilahkan pilih bantuan yang bisa aku berikan : ")
    print("1. Tentang hari ini")
    print("2. Mencatat jadwal agenda harianmu")
    print("3. Keluar")
    
#Tentang Hari ini
def show_menu_1():
    print("\nApa yang kamu ingin tahu tentang hari ini ?")
    print("1. Hari ini hari apa?")
    print("2. Hari ini tanggal berapa")
    print("3. Keluar")
    
#Variabel Hari dan Tanggal
x = datetime.datetime.now()
tanggal = x.strftime("%d-%m-%Y")
hari = x.strftime("%A")
hari_indo = {"Sunday" : "Minggu", "Monday" : "Senin", "Tuesday" : "Selasa", "Wednesday" : "Rabu", "Thursday" : "Kamis", "Friday" : "Jumat", "Saturday" : "Sabtu"}
jawaban_tanggal = ["Hari ini hari {}".format(hari_indo[hari]), "tanggal {}".format(tanggal)]

def show_menu_2():
    print("/nApa agendamu ?")
    print("1. Daftar agenda")
    print("2. Menambah agenda")
    print("3. Keluar")
    
daftar_agenda = []

#tampilkan agenda
def tampilkan_agenda():
    if not daftar_agenda:
        print("Daftar Agendamu masih kosong, {}!\n Coba kamu tambah agendamu dulu".format(nama))
    else:
        print("Total agenda ada {}, ini adalah agendamu : ".format(len(daftar_agenda)))
        print("=========================================")
        for agenda in daftar_agenda:
            print("Daftar Kegiatan : ", agenda["Kegiatan"])
            print("Dilakukan tanggal : ", agenda ["Tanggal"])
            print("Dilakukan jam : ", agenda["Jam"])
            print("------------------------------------------")
            
#tambah agenda
def tambah_agenda():
    Kegiatan = input("Apa rencana kegiatanmu: ")
    Tanggal = input("Tanggal berapa rencananya (format DD-MM_YYYY) : ")
    Jam = input("Jam berapa (HH:MM): ")
    agenda = {
        "Kegiatan": Kegiatan,
        "Tanggal" : Tanggal,
        "Jam" : Jam
    }
    daftar_agenda.append(agenda)
    print("+=========================+")
    print("Agendamu telah berhasil ditambahkan")
    print("+=========================+")
    
#create agenda
def create_agenda():
    f = open("Daftar agenda.txt", "w")
    f.write("Total agenda ada {}, ini rinciannya : \n".format(len(daftar_agenda)))
    f.write("-------------------------------------------------")
    for agenda in daftar_agenda:
        f.write("\nRencana kegiatanmu : {} \nDilakukan pada tanggal {}\nJam : {}\n".format(agenda["Kegiatan"], agenda ["Tanggal"], agenda["Jam"]))
        f.write("----------------------------------------------------")
    f.close()
while True:
    show_menu()
    menu_awal = input("Pilih hanya angka aja ya: \n")
    
    if menu_awal == "1":
        #Tentang hari ini
        while True:
            show_menu_1()
            menu_hari_ini = input("Pilih hanya angka aja ya: \n")
            
            if menu_hari_ini == "1":
                print("=========================================================")
                print(jawaban_tanggal [0])
                print("=========================================================")
            elif menu_hari_ini == "2":
                print("======================================")
                print(jawaban_tanggal [1])
                print("======================================")
            elif menu_hari_ini == "3":
                print("\nYasuda kalo gitu")
                print("======================================")
                break
            else:
                print("======================================")
                print("Menu tidak tersedia")
    
    elif menu_awal == "2":
        #Agenda
        while True:
            show_menu_2()
            menu_agenda = input("Pilih hanya angka aja ya: \n")
            
            if menu_agenda == "1":
                tampilkan_agenda()
            elif menu_agenda == "2":
                tambah_agenda()
            elif menu_agenda == "3":
                print("\nYasuda kalo gitu")
                print("======================================")
                create_agenda ()
                break
            else:
                print("======================================")
                print("Menu tidak tersedia")

    elif menu_awal == "3":
        print("\nOke, sampe jumpa")
        print("======================================")
        break
    
    else:
        print("======================================")
        print("Menu tidak tersedia")
```
