```python
daftar = []

def menu():
    print("\n--- Menu ---")
    print("1. Daftar Kontak")
    print("2. Tambah Kontak")
    print("3. Keluar")
    

def daftar_kontak(daftar):
    if len(daftar) == 0:
        print("Maaf, belum ada kontak yang terdaftar")
    else :
        print("Daftar Kontak : ")
        for i in daftar:
            print("Nama : {}".format(i['Nama']))
            print("No Telepon : {}".format(i['No_Telp']))

def tambah_kontak(daftar):
    kontak_baru = {'Nama' : '', 'No_Telp' : ''}
    nama_baru = input("Nama : ")
    kontak_baru['Nama'] = nama_baru
    no_baru = input("No Telepon : ")
    kontak_baru['No_Telp'] = no_baru
    daftar.append(kontak_baru)
    print("Kontak berhasil ditambahkan")
    

#Pembuka
print("Selamat Datang")
while True:
    menu()
    x = input("Pilih Menu : ")
    if x == '1': 
        daftar_kontak(daftar)
    elif x == '2':
        tambah_kontak(daftar)
    elif x == '3':
        print("Program selesai, Sampai Jumpa")
        break
    else:
        print("Menu tidak tersedia, silakan coba lagi")
```
